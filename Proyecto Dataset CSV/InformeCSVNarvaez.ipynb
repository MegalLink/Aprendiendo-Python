## Google Play Store Apps CSV
El siguiente data set fue obtenido de [Kaggle](https://www.kaggle.com/lava18/google-play-store-apps)
### Resumen

Si bien muchos conjuntos de datos públicos (en Kaggle y similares) proporcionan datos de la App Store de Apple, no hay muchos conjuntos de datos equivalentes disponibles para las aplicaciones de Google Play Store en la web. Google Play Store utiliza sofisticadas técnicas modernas (como la carga dinámica de páginas) utilizando JQuery, lo que hace que el scraping sea más desafiante.

El objetivo de este dataset es desmostrar que los datos de las aplicaciones de Play Store tienen un enorme potencial para impulsar al éxito a las empresas de creación de aplicaciones. Se pueden extraer conocimientos prácticos para que los desarrolladores trabajen y capturen el mercado de Android.

### Contenido
El dataset contiene el nombre de la aplicación, la categoría, el rating, el número de comentarios de los usuarios, el peso de la aplicación, número de veces instalado, los generos, la última vez que se actualizo, la versión actual y para que version de android esta diseñada.




### Cargando CSV


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset_path = "googleplaystore.csv"
dataf = pd.read_csv(dataset_path)
dataf.Installs=pd.to_numeric(dataf.Installs)
dataf.Price=pd.to_numeric(dataf.Price)
dataf.Rating=pd.to_numeric(dataf.Rating)
dataf.drop_duplicates()
print(dataf)
def grafica(df,titulo,xlabel,ylabel,color,x_param,y_param):
    g=df.plot(title=titulo,kind='bar',x=x_param,y=y_param,color=color)
    g.set_xlabel(xlabel)
    g.set_ylabel(ylabel)
    plt.show()
  
```

### Preguntas

1. Cual es el costo de las aplicaciones de paga con mayor rating?


```python
df=dataf.sort_values(by=['Type','Rating'],ascending=[False, False])[['App','Rating','Type','Price']].head(10)
print(df)

grafica(df,"Precio aplicaciones de paga con mayor Raiting","Aplicacion","Precio","blue","App","Price")
```

2. Cuales son las aplicaciones gratuitas con mayor Rating?

```python
df=dataf.sort_values(by=['Rating','Price'],ascending=[False,True]).head(10)[['App','Rating','Price','Type']]
print(df)
grafica(df,"Aplicaciones gratuitas con mayor Raiting","Aplicacion","Rating","green","App","Rating")
```

3. Cuales son las aplicaciones gratuitas más descargadas?

```python
df=dataf.sort_values(by=['Installs','Type'],ascending=[False,False]).head(10)[['App','Installs','Price']]
print(df)
grafica(df,"Aplicaciones gratuitas mas descargadas","Aplicacion","Descargas","red","App","Installs")
```

4. Cuales son las aplicaciones de paga mas descargadas?


```python
df=dataf.sort_values(by=['Type','Installs'],ascending=[False,False]).drop_duplicates().head(10)[['App','Installs','Price','Type']]
print(df)
grafica(df,"Aplicaciones de paga mas descargadas","Aplicacion","Descargas","yellow","App","Installs")
```

5. Numero de descargas de las aplicaciones de paga mas caras?

```python
df=dataf.sort_values(by=['Type','Installs'],ascending=[False,False]).drop_duplicates().head(10)[['App','Installs','Price']]
print(df)
grafica(df,"Descargas de aplicaciones más caras","Precio","Descargas","orange","Price","Installs")
```

6. Cuales son las categorias con mas aplicaciones?

```python

serie=dataf.groupby("Category")["Category"].count().sort_values(ascending=False)
print(serie)
grafica(serie,"Aplicaciones desarrolladas por Categoria","Categoria","Cantidad","blue","index","Category")
```

7. Cuales son los generos mas descargados?

```python
serie=dataf.groupby(["Category"])["Installs"].sum().sort_values(ascending=False)
print(serie)
grafica(serie,"Descargas por categoria","Categoria","Descargas","blue","index","Category")
```

8. Cuanto recauda cada genero?

```python
serie=dataf.groupby(["Category"])["Price"].sum().sort_values(ascending=False)
print(serie)
grafica(serie,"Recaudacion por categoria","Categoria","Recaudacion","blue","index","Category")
```

9. Con que version de android como base se desarrollan la mayoria de aplicaciones?

```python
serie=dataf.groupby(["Android Ver"])["Installs"].count().sort_values(ascending=False)
print(serie)
grafica(serie,"Version de desarrollo de aplicaciones","Version","# Aplicaciones","red","index","Category")
```

10. Cual es la categoria con mas reviews?


```python
serie=dataf.groupby(["Category"])["Reviews"].sum().sort_values(ascending=False)
print(serie)
grafica(serie,"Reviews por categoria","Categoria","# Reviews","red","index","Category")
```

11. Cuanto dinero mueven en total las aplicaciones moviles?

```python
serie=dataf["Price"].sum()
print("Total dinero en todas las aplicaciones",serie)
```

12. Cantidad de aplicaciones gratuitas vs Cantidad de aplicaciones de paga

```python
serie=dataf.groupby(["Type"])["Type"].count().sort_values(ascending=False)
print(serie)
grafica(serie,"Reviews por categoria","Categoria","# Reviews","yellow","index","Category")
```

13. Cuales son los juegos mas vendidos?

```python
df=dataf.loc[(dataf['Category']=="GAME") & (dataf['Type'] == "Paid")][["App","Installs","Type"]].sort_values('Installs',ascending=False).drop_duplicates().head(10)
print(df)
grafica(df,"Juegos mas vendidos","Juego","Descargas","DarkBlue","App","Installs")
```

14. Cuales son los juegos gratuitos mas descargados


```python
df=dataf.loc[(dataf['Category']=="GAME") & (dataf['Type'] == "Free")][["App","Installs","Type"]].sort_values('Installs',ascending=False).drop_duplicates().head(10)
print(df)
grafica(df,"Juegos mas vendidos","Juego","Descargas","DarkOrange","App","Installs")
```

15. Aplicaciones de comunicacion mas descargadas


```python
df=dataf.loc[(dataf['Category']=="COMMUNICATION") ][["App","Installs","Type"]].sort_values('Installs',ascending=False).drop_duplicates().head(10)
print(df)
grafica(df,"Aplicaciones de comunicacion mas descargadas","Aplicacion","Descargas","DarkRed","App","Installs")
```

16. Aplicaciones sociales mas descargadas


```python
df=dataf.loc[(dataf['Category']=="SOCIAL") ][["App","Installs","Type"]].sort_values('Installs',ascending=False).drop_duplicates().head(10)
print(df)
grafica(df,"Aplicaciones de comunicacion mas descargadas","Aplicacion","Descargas","Gray","App","Installs")
```

17. Hacia que publico esta orientado la mayoria de aplicaciones?


```python
serie=dataf.groupby("Content Rating")["Content Rating"].count().sort_values(ascending=False)
print(serie)
grafica(serie,"Orientacion hacia el publico de las aplicaciones","Categoria","Cantidad","Pink","index","Category")
```

18. Cual es el peso de las aplicaciones mas descargadas?

```python
df=dataf.loc[(dataf['Size']!="Varies with device") ].sort_values(by=['Installs','Size'],ascending=[False, False])[['App','Installs','Type','Size']].drop_duplicates().head(10)
print(df)

grafica(df,"Peso de las aplicaciones mas descargadas","Peso","Descargas","blue","Size","Installs")
```

19. Cuantos dinero recaudan los juegos mas vendidos?

```python
df=dataf.loc[(dataf['Category']=="GAME") & (dataf['Type'] == "Paid")][["App","Installs","Price"]].sort_values('Installs',ascending=False).drop_duplicates().head(10)

installs=df["Installs"]
price=df["Price"]
total=installs*price
#print(type(total))
df['Total']=total
print(df)
grafica(df,"Dinero recaudado en ventas por videojuego","Juego","Dinero recaudado","Teal","App","Total")
```

20. Dinero recaudado por las aplicaciones mas vendidas?


```python
df=dataf.loc[(dataf['Type'] == "Paid")][["App","Installs","Price"]].sort_values('Installs',ascending=False).drop_duplicates().head(10)

installs=df["Installs"]
price=df["Price"]
total=installs*price
#print(type(total))
df['Total']=total
print(df)
grafica(df,"Dinero recaudado en ventas por Aplicaciones","Aplicaciones","Dinero recaudado","Brown","App","Total")
```
