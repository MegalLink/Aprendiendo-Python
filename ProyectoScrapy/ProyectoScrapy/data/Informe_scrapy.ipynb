### Resumen 
Scrapy es una plataforma colaborativa de código libre que corre en Python para extraer datos de páginas web usado para una serie de aplicaciones como minería de datos, procesamiento de información o registro histórico.

Scrapy tiene las siguientes características: 

- Rápida y poderosa: Escribes las reglas para extraer los datos y scrapy hace el resto.
- Fácilmente extensible: Dada su configuración, puede generara nueva funcionalidad sin tener que modificar el código fuente.
- Portable y Pythonico: Esta escrito en Python y puede correr en Linux, Windows, Mac y BSD.

Gracias a esta herramienta se desarrollo el proyecto obteniendo la información de una página donde normalmente veo series de forma gratuita que estan contenidas en varias plataformas como Prime, Netflix, etc. Los datos fueron extraidos de [enlace en línea](https://seriesflix.to)
### Contenido
Los datos extraídos son los siguientes:
actores,estado,fecha_estreno,generos,puntaje,temporadas,titulo



| Campos  | Descripción  |
|:--|---|
|Actores   | Lista del reparto de la serie   |
|Estado   | En emisión o finalizado  |
|Fecha de estreno   | Año cuando se estreno la serie  |
|Generos   | Lista de generos que contiene la serie  |
|Puntaje   | Puntuación dada por el usuario sobre 5  |
| Temporadas  | Número de temporadas que contiene la serie  |
| Titulo  | Nombre de la serie  |


### Cargar CSV

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataset_path = "series_data.csv"
dataf = pd.read_csv(dataset_path)
dataf.drop_duplicates()
print(dataf)
```

#### Series con estado Finalizado vs en Emision


```python
serie=dataf.groupby("estado")["estado"].count().sort_values(ascending=False)
print(serie)
plt.pie(serie, labels=["Emision","Finalizada"])
plt.show()
```

#### Cantidad de series con varias categorias


```python
drama=dataf.loc[(dataf['generos'].str.contains("Drama", case=False))].count()
cf=dataf.loc[(dataf['generos'].str.contains("Ciencia ficción", case=False))].count()
comedia=dataf.loc[(dataf['generos'].str.contains("Comedia", case=False))].count()
misterio=dataf.loc[(dataf['generos'].str.contains("Misterio", case=False))].count()

plt.pie([drama.generos,cf.generos,comedia.generos,misterio.generos], labels=["Drama","Ciencia ficción","Comedia","Misterio"])
plt.show()
```

#### Series con mayor puntaje

```python
df=dataf.sort_values(by=['puntaje'],ascending=[False])[['titulo','puntaje','temporadas']].head(10)
print(df)
g=df.plot(title="Series con mayor puntaje",kind='bar',x="titulo",y="puntaje",color="teal")
g.set_xlabel("Titulo")
g.set_ylabel("Puntaje")
plt.show()
```

#### Series con mas temporadas

```python
df=dataf.sort_values(by=['temporadas'],ascending=[False])[['titulo','puntaje','temporadas']].head(10)
print(df)
g=df.plot(title="Series con mas temporadas",kind='bar',x="titulo",y="temporadas",color="red")
g.set_xlabel("Titulo")
g.set_ylabel("Temporadas")
plt.show()
```

Cantidad de series estrenadas por años

```python
serie=dataf.groupby("fecha_estreno")["fecha_estreno"].count().sort_values(ascending=False)
print (serie)
fig, ax = plt.subplots()

ax.set_ylabel('Fecha estreno')
ax.set_xlabel('Estreno series')
ax.set_title('Series estrenadas por año')
plt.bar(serie.index, serie.values)
plt.show()
```

#### Series con Mayor puntaje del 2020

```python
df=dataf.sort_values(by=['puntaje'],ascending=[False])[['titulo','puntaje','fecha_estreno']]
df=df.loc[(df['fecha_estreno']==2020) ].head(20)
print(df)
g=df.plot(title="Series con mayor puntaje",kind='bar',x="titulo",y="puntaje",color="yellow")
g.set_xlabel("Titulo")
g.set_ylabel("Puntaje")
plt.show()
```
